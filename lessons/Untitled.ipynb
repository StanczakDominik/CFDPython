```python
import numpy as np
from matplotlib import pyplot as plt
import time, sys
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

def run(n, NT, sigma, c, color):
    y, dy = x, dx = np.linspace(0,2,n, retstep=True)
    X, Y = np.meshgrid(x,y)
    dt = dx*sigma
    u = np.ones((n,n))
    u[(.5 <= X) * (X <= 1) * (.5 <= Y) * (Y <= 1)] = 2
    
    fig = plt.figure(figsize=(11,7), dpi=100)
    ax = fig.gca(projection='3d')
#     surf = ax.plot_surface(X,Y,u)
    
    for t in range(NT):
        u_back_x = u[:-1,1:]#np.roll(u, -1, axis=0)
        u_back_y = u[1:,:-1]#np.roll(u, -1, axis=1)
        u_center = u[1:,1:]
                           
        u[1:,1:] = u_center - c*dt/dx*(u_center - u_back_x) - c*dt/dy*\
                (u_center-u_back_y)   
            
#     plt.plot(x,u, color+"-", label="computational")
#     plt.xlim(0,2*np.pi)
#     plt.plot(x,ufunc(NT*dt, x, nu), color+"--", label="theory")

run(81,100, .2, 1, "b")
```
