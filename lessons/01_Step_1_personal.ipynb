```python
import numpy as np
from matplotlib import pyplot as plt
import time, sys
%matplotlib inline

def run(nx, T, NT, c, color):
    x, dx = np.linspace(0,2,nx, retstep=True)
    dt = T/(NT) #not NT+1...

    u = np.ones(nx)
    u[int(.5/dx) : int(1/dx+1)] =2
    plt.plot(x, u, color+"--")
    for t in range(NT):
        u[1:] = u[1:] - c*dt/dx * (u[1:] - u[:-1])
    plt.plot(x,u, color+"-")
```

# 1D linear convection
$$\frac{\partial u}{\partial t} + c \frac{\partial u}{\partial x} = 0$$

```python
run(nx=81, T= 25*0.025, NT=25, c=1, color="b")
run(nx=41, T= 25*0.025, NT=25, c=1, color="r")
run(nx=21, T= 25*0.025, NT=25, c=1, color="g")
```

```python
run(nx=81, T= 25*0.025, NT=25, c=1, color="b")
run(nx=81, T= 25*0.025, NT=250, c=1, color="r")
run(nx=81, T= 25*0.025, NT=2500, c=1, color="g")
run(nx=81, T= 25*0.025, NT=25000, c=1, color="c")
run(nx=81, T= 25*0.025, NT=250000, c=1, color="v")
```
