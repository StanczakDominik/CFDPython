# 1D diffusion

$$\frac{\partial u}{\partial t} = \nu \frac{\partial^2 u}{\partial x^2}$$

```python
import numpy as np
from matplotlib import pyplot as plt
import time, sys
%matplotlib inline

def run(nx, T, nu, sigma, color):
    x, dx = np.linspace(0,2,nx, retstep=True)
    dt = sigma*dx**2/nu
    NT = int(T/dt)
    u = np.ones(nx)
    u[int(.5/dx) : int(1/dx+1)] =2
    plt.plot(x, u, color+"--")
    for t in range(NT):
        u[1:-1] = u[1:-1] + nu*dt/dx**2 * (u[2:] - 2*u[1:-1] + u[:-2])   
    plt.plot(x,u, color+"-")
run(41,0.03333333333333335, 0.3, 0.2, "b")
run(41,0.03333333333333335, 0.6, 0.2, "g")
run(41,0.03333333333333335, 1.6, 0.2, "r")
run(41,0.03333333333333335, 8, 0.2, "k")
run(41,0.03333333333333335, 30, 0.2, "y")
```
