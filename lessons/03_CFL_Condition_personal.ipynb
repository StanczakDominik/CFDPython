```python
import numpy as np
from matplotlib import pyplot as plt
import time, sys
%matplotlib inline

def run(nx, T, NT, color, nonlinear=False):
    x, dx = np.linspace(0,2,nx, retstep=True)
    dt = T/(NT) #not NT+1...

    u = np.ones(nx)
    u[int(.5/dx) : int(1/dx+1)] =2
    plt.plot(x, u, color+"--")
    for t in range(NT):
        
        if nonlinear:
            c = u[1:]
        else:
            c=1
        u[1:] = u[1:] - c*dt/dx * (u[1:] - u[:-1])   
    plt.plot(x,u, color+"-")
```

```python
run(41,20*.025,20,"b")
run(61,20*.025,20,"g")
run(71,20*.025,20,"y")
run(81,20*.025,20,"c")
```

```python
run(82,20*.025,20,"b")
run(83,20*.025,20,"g")
run(84,20*.025,20,"r")
run(85,20*.025,20,"c")
```

# CFL condition

$$\sigma = \frac{u \Delta t}{\Delta x} \leq \sigma_{max}$$

Nondimensionalized wave velocity, has to be less than $sigma_max$ (depends on discretization used)

```python
def run_CFL(nx, T, color):
    x, dx = np.linspace(0,2,nx, retstep=True)
    c=1
    sigma = .5
    dt = sigma*dx
    NT = T/dt

    u = np.ones(nx)
    u[int(.5/dx) : int(1/dx+1)] =2
    plt.plot(x, u, color+"--")
    for t in range(NT):
        u[1:] = u[1:] - c*dt/dx * (u[1:] - u[:-1])   
    plt.plot(x,u, color+"-")
```

```python
for NX, color in [(41, "b"), (121, "g"), (181, "y"), (281,"c")]:
    run(41,20*.025,20,"b")
    run(61,20*.025,20,"g")
    run(71,20*.025,20,"y")
    run(81,20*.025,20,"c")
```

Fulfill Courant condition and wave remains stable
