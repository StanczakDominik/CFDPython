```python
import numpy as np
from matplotlib import pyplot as plt
import time, sys
%matplotlib inline

def run(nx, T, NT, color):
    x, dx = np.linspace(0,2,nx, retstep=True)
    dt = T/(NT) #not NT+1...

    u = np.ones(nx)
    u[int(.5/dx) : int(1/dx+1)] =2
    plt.plot(x, u, color+"--")
    for t in range(NT):
        u[1:] = u[1:] - u[1:]*dt/dx * (u[1:] - u[:-1])   
    plt.plot(x,u, color+"-")
```

```python
run(nx=41, T=20*0.025, NT = 20, color="b")
run(nx=41, T=20*0.025, NT = 40, color="g")
run(nx=41, T=20*0.025, NT = 80, color="r")
run(nx=41, T=20*0.025, NT = 8000, color="c")
```

# 1D *Non*linear convection

$$\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} = 0$$

Discretizes as

$$u_i^{n+1} = u_i^n - u_i^n \frac{\Delta t}{\Delta x} (u_i^n - u_{i-1}^n)$$

Seems to diffuse more and more with increasing NT, decreasing timestep. Why?
